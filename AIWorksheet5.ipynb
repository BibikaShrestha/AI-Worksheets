```python
def cost_function(X, Y, W):
    X = np.array(X, dtype=float)
    Y = np.array(Y, dtype=float).reshape(-1, 1)
    W = np.array(W, dtype=float).reshape(-1, 1)

    n = len(Y)
    Y_pred = X @ W
    error = Y_pred - Y
    cost = (1 / (2 * n)) * np.sum(error ** 2)

    return cost
```

```python

def gradient_descent(X, Y, W, alpha, iterations):
    Y = Y.reshape(-1, 1)
    m = len(Y)
    cost_history = []

    for _ in range(iterations):
        # Prediction
        Y_pred = X @ W

        # Error
        loss = Y_pred - Y

        # Gradient
        dw = (1 / m) * (X.T @ loss)

        # Update weights
        W = W - alpha * dw

        # Store cost
        cost = cost_function(X, Y, W)
        cost_history.append(cost)

    return W, cost_history
```

```python


def rmse(Y, Y_pred):
    Y = Y.reshape(-1, 1)
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

To - Do - 9:

```python
def r2(Y, Y_pred):
    Y = Y.reshape(-1, 1)
    mean_y = np.mean(Y)

    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)

    return 1 - (ss_res / ss_tot)
```

To - Do - 10:

We will define a function that:
1. Loads the data and splits it into training and test sets.
2. Prepares the feature matrix (X) and target vector (Y).
3. Defines the weight matrix (W) and initializes the learning rate and number of iterations.
4. Calls the gradient descent function to learn the parameters.
5. Evaluates the model using RMSE and R2

```python
def main():
    # Load data
    df = pd.read_csv('/content/drive/MyDrive/student.csv')

    # Handle missing values (as you did)
    missing_info = df.isnull().sum() / len(df) * 100
    for column in df.columns:
        if missing_info[column] > 10:
            df[column].fillna(df[column].median(), inplace=True)
        else:
            df.dropna(subset=[column], inplace=True)

    # Features and target
    X = df[['Math', 'Reading']].values
    y = df['Writing'].values

    # Split
    X_train, X_test, y_train, y_test = train_test_split_scratch(X, y, test_size=0.25)

    # Initialize weights
    W_init = np.zeros((X_train.shape[1], 1))
    alpha = 0.00001
    iterations = 1000

    # Train
    W_final, cost_history = gradient_descent(X_train, y_train, W_init, alpha, iterations)

    # Predict
    y_pred_test = X_test @ W_final

    # Evaluate
    print("Final Weights:\n", W_final)
    print("RMSE:", rmse(y_test, y_pred_test))
    print("R² Score:", r2(y_test, y_pred_test))

    return W_final, cost_history

if __name__ == "__main__":
    main()
```
